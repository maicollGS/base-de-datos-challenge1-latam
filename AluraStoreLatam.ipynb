#1. Análisis de facturación



##Ingreso total por cada tienda

En este primer análisis, se calcula el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
def ingreso_tienda(dataset):
    total = dataset['Precio'].sum()
    print(f"Ingreso total de la tienda: ${total:,.2f}")
```

####Ingresos tienda 1

```python
ingreso_total_tienda1 = ingreso_tienda(tienda)
ingreso_total_tienda1
```

####Ingresos tienda 2

```python
ingreso_total_tienda2 = ingreso_tienda(tienda2)
ingreso_total_tienda2
```

####Ingresos tienda 3

```python
ingreso_total_tienda3 = ingreso_tienda(tienda3)
ingreso_total_tienda3
```

####Ingresos tienda 4

```python
ingreso_total_tienda4 = ingreso_tienda(tienda4)
ingreso_total_tienda4
```

```python
def total_Agrupado(dato1, dato2, dato3, dato4, Columname):
  series = {'tienda1' : dato1,
            'tienda2' : dato2,
            'tienda3' : dato3,
            'tienda4' : dato4}
  return pd.Series(series, name = Columname)
```

```python
Totales_Tiendas = total_Agrupado(ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4, 'Ingresos').astype(int)

Totales_Tiendas
```

# 2. Ventas por categoría

En este se calcula la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

#####Conteo categorias Tienda 1

```python
conteo_categorias_Tienda1 = tienda['Categoría del Producto'].value_counts().sort_values(ascending=False).head(3)
print(conteo_categorias_Tienda1)
```

####Conteo categorias Tienda 2

```python
conteo_categorias_Tienda2 = tienda2['Categoría del Producto'].value_counts().sort_values(ascending=False).head(3)
print(conteo_categorias_Tienda2)
```

####Conteo categorias Tienda 3

```python
conteo_categorias_Tienda3 = tienda3['Categoría del Producto'].value_counts().sort_values(ascending=False).head(3)
print(conteo_categorias_Tienda3)
```

####Conteo categorias Tienda 3

```python
conteo_categorias_Tienda4 = tienda4['Categoría del Producto'].value_counts().sort_values(ascending=False).head(3)
print(conteo_categorias_Tienda4)
```

# 3. Calificación promedio de la tienda


En este paso, se deben calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.
